```python
#step 2: importing libraries 
import matplotlib.pyplot as mpl
import seaborn as sb
import pandas as pd
import numpy as np
```

```python
#step 3: importing lending club data from git hub personal repository
df= pd.read_csv('https://raw.githubusercontent.com/Skalvis/DSA_EDA/master/loans_2007.csv')
```

```python
#step 4: printing histogram for loan amount (x= loan value; y= loan value reoccurance frequency)
mpl.hist(df['loan_amnt'])
```

```python
#step 5: prints a bar chart with sum of loan amount per purpose
purpose = pd.DataFrame() #new data frame for unique sums
names=['purpose_car','purpose_credit_card', 'purpose_debt_consolidation', 'purpose_educational','purpose_home_improvement',
       'purpose_house', 'purpose_major_purchase', 'purpose_medical', 'purpose_moving', 'purpose_other', 'purpose_renewable_energy',
       'purpose_small_business', 'purpose_vacation', 'purpose_wedding'] #all purpose names within one list variable

#all sums of purposes within one list variable
sums=[df.query('purpose_car == 1')['loan_amnt'].sum(),df.query('purpose_credit_card == 1')['loan_amnt'].sum(),
     df.query('purpose_debt_consolidation == 1')['loan_amnt'].sum(), df.query('purpose_educational == 1')['loan_amnt'].sum(),
     df.query('purpose_home_improvement == 1')['loan_amnt'].sum(),df.query('purpose_house == 1')['loan_amnt'].sum(),
     df.query('purpose_major_purchase == 1')['loan_amnt'].sum(), df.query('purpose_medical == 1')['loan_amnt'].sum(),
     df.query('purpose_moving == 1')['loan_amnt'].sum(), df.query('purpose_other == 1')['loan_amnt'].sum(),
     df.query('purpose_renewable_energy == 1')['loan_amnt'].sum(),df.query('purpose_small_business == 1')['loan_amnt'].sum(),
     df.query('purpose_vacation == 1')['loan_amnt'].sum(),df.query('purpose_wedding == 1')['loan_amnt'].sum()]

#assigned to new df
purpose['Purpose'] = names
purpose['Sums'] = sums

#rotating x axis names
mpl.xticks(rotation='vertical')
#bar diagram
mpl.bar(purpose['Purpose'],purpose['Sums'])
```

```python
#step 6: boxplot for "loan amount", printing median + min + max + Q1+ Q3 + outliers
sb.boxplot(x=df['loan_amnt'])
df['loan_amnt'].min(), df['loan_amnt'].max()
#as seen from the boxplot diagram, min value is "0" without any lower outliers, max value is not within Q1-Q3, it is outlier 
#value of "35000". Q1 is middle value between min and mediam ="5000", mediam = "10000", Q3 is middle value between mediam and 
#max which is = "15000". 
```
